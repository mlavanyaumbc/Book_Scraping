```python
import requests
import sqlite3
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = 'https://books.toscrape.com/'
db_name = 'Books.db'
table_name = 'Top_20'
df = pd.DataFrame(columns=["image_container","Star-rating","title","product_price","instock availability"])
```

```python
# Send a request to the website and parse the HTML content
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
# scrape all books
for book in soup.select('.product_pod'):

    # Extract data points
    image_container = book.select_one('.thumbnail').get('src')  # Image URL
    star_rating = book.select_one('.star-rating').get('class')[1]  # Rating
    title = book.select_one('h3 a').get('title')  # Title of the book
    product_price = book.select_one('.price_color').text  # Price
    instock_availability = book.select_one('.availability').text.strip()  # Availability status

    # Create temp DataFrame and concatenate
    temp_df = pd.DataFrame([{
        "image_container": image_container,
        "Star-rating": star_rating,
        "title": title,
        "product_price": product_price,
        "instock availability": instock_availability
    }])
    df = pd.concat([df, temp_df], ignore_index=True)

# Convert star ratings to numeric values for sorting (optional)
df['Star-rating'] = df['Star-rating'].map({"One": 1, "Two": 2, "Three": 3, "Four": 4, "Five": 5})

# Sort by 'Star-rating' descending and select the top 20
top_20_books = df.sort_values(by='Star-rating', ascending=False).head(20)


```
